## Plotting spectra of cloud with different abundances

### Solar abundance

```python
cloud_solar = pd.read_table('cloud_solar.con', header=0)
cloud_solar
```

```python
cloud_solar.shape
```

```python
cloud_solar.columns
```

```python
cloud_solar_sub = cloud_solar.loc[np.float64(cloud_solar['#Cont  nu'])>1e-1]
cloud_solar_sub = cloud_solar_sub.loc[np.float64(cloud_solar_sub['#Cont  nu'])<1e3]
cloud_solar_sub.shape
```

```python
plt.figure(figsize=(12, 5))

plt.loglog(cloud_solar_sub['#Cont  nu'], cloud_solar_sub['total'], c='darkorange', linewidth=0.5)

plt.xlabel(r'Wavelength [$\mu$m]')
plt.ylabel(r'Flux [erg s$^{-1}$]')

plt.title('Spectra of a Cloud with Solar Abundances')

# plt.savefig('spec_cloud_solar.pdf')

plt.show()
```

## Gas Temperature

```python
cloud_solar_ovr = pd.read_table('cloud_solar.ovr')
cloud_solar_ovr
```

```python
cloud_solar_ovr.shape
```

```python
cloud_solar_ovr.columns
```

#### Splitting

```python
idx_solar = cloud_solar_ovr.loc[cloud_solar_ovr['#depth']=='###########################'].index[0]
type(idx_solar)
```

```python
solar_ovr_1 = cloud_solar_ovr[:idx_solar]
solar_ovr_1.shape
```

```python
solar_ovr_1  # index finishes in 115
```

```python
solar_ovr_2 = cloud_solar_ovr[idx_solar+1:]
solar_ovr_2.shape
```

```python
solar_ovr_2  # index starts in 117
```

```python
# verify splitting is correct, should be true
print(solar_ovr_1.shape[0] + solar_ovr_2.shape[0] + 1 == cloud_solar_ovr.shape[0])
```

```python
cloud_PNe_ovr = pd.read_table('cloud_PNe.ovr')
cloud_PNe_ovr
```

```python
cloud_PNe_ovr.shape
```

```python
cloud_PNe_ovr.columns
```

```python
idx_pne = cloud_PNe_ovr.loc[cloud_PNe_ovr['#depth']=='###########################'].index[0]
type(idx_pne)
```

```python
PNe_ovr_1 = cloud_PNe_ovr[:idx_pne]
PNe_ovr_1.shape
```

```python
PNe_ovr_1  # index finishes in 150
```

```python
PNe_ovr_2 = cloud_PNe_ovr[idx_pne+1:]
PNe_ovr_2.shape
```

```python
PNe_ovr_2  # index starts in 152
```

#### Swithing to float

```python
solar_ovr_1['#depth'] = solar_ovr_1['#depth'].astype('float64')
solar_ovr_2['#depth'] = solar_ovr_2['#depth'].astype('float64')

PNe_ovr_1['#depth'] = PNe_ovr_1['#depth'].astype('float64')
PNe_ovr_2['#depth'] = PNe_ovr_2['#depth'].astype('float64')
```

### Plotting

```python
plt.figure(figsize=(12, 5))

plt.loglog(solar_ovr_1['#depth'], solar_ovr_1['Te'], c='darkorange', linewidth=1.0, label='Solar, 1')
plt.loglog(solar_ovr_2['#depth'], solar_ovr_2['Te'], c='darkred', linewidth=1.0, label='Solar, 2')

plt.xlabel(r'Depth [cm]')
plt.ylabel(r'Gas Temperature [K]')

plt.legend(prop={'size': 12})

# plt.savefig('spec_Tgas_solar.pdf')

plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.loglog(PNe_ovr_1['#depth'], PNe_ovr_1['Te'], c='seagreen', linewidth=1.0, label='PNe, 1')
plt.loglog(PNe_ovr_2['#depth'], PNe_ovr_2['Te'], c='royalblue', linewidth=1.0, label='PNe, 2')

plt.xlabel(r'Depth [cm]')
plt.ylabel(r'Gas Temperature [K]')

plt.legend(prop={'size': 12})

# plt.savefig('spec_Tgas_pne.pdf')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

plt.plot(solar_ovr_1['#depth'], solar_ovr_1['Te'], c='darkred', linewidth=1.0, label='Solar, iter 1')
plt.plot(solar_ovr_2['#depth'], solar_ovr_2['Te'], c='darkorange', linewidth=1.0, label='Solar, iter 2')

plt.plot(PNe_ovr_1['#depth'], PNe_ovr_1['Te'], c='royalblue', linewidth=1.0, label='PNe, iter 1')
plt.plot(PNe_ovr_2['#depth'], PNe_ovr_2['Te'], c='seagreen', linewidth=1.0, label='PNe, iter 2')

ax.set_yscale('log')

plt.xlabel(r'Depth [cm]')
plt.ylabel(r'Gas Temperature [K]')

plt.legend(prop={'size': 12})

# plt.savefig('spec_Tgas.pdf')

plt.show()
```
